# Projet d'étude de cas sur l'apprentissage automatique de la classification

### Data transforms and machine learning

```python
# Load dataset
from pandas import read_csv
from sklearn.preprocessing import Normalizer
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import KFold, cross_val_score
filename = 'sonar.csv'
labels = ['preg', 'plas', 'pres', 'skin', 'test', 'mass', 'pedi', 'age', 'class'] 
dataframe = read_csv(filename, names = labels)
data_values = dataframe.values # Retrieve values from the dataframe
# Separate array into input and output components
X = data_values[  : , 0: -1] # data_values[  : , 0 :-1]
Y = data_values[  : , -1] # data_values[  : , -1] 
#nachine learning
scaler = Normalizer().fit(X)
normalizedX = scaler.fit_transform(X) 


```

# K= 10

## Classification Accuracy

```python
seed = 7
split = 10
# Create 10-folds
kfold = KFold(n_splits = split, random_state=seed, shuffle=True)
model = LogisticRegression()
metric = 'accuracy'
results = cross_val_score(model, X, Y, cv=kfold, scoring = metric)
# Prediction -> Success rate (Accuracy)
print("Accuracy: ",(results.mean()*100.0), results.std()*100.0)
```

## Logarithmic Loss

```python

metric = 'neg_log_loss'
results = cross_val_score(model, X, Y, cv=kfold, scoring = metric)
# Prediction -> Success rate (Accuracy)
print("Log Loss: ",(results.mean()), results.std())
```

```python
'''
l'ecart de la valeur numérique entre l'étiquette de probabilité présumée et la vrai est faible:
on en déduit que la prediction est efficace
'''
```

## Area Under ROC Curve (AUC)

```python

metric = 'roc_auc'
results = cross_val_score(model, X, Y, cv=kfold, scoring = metric)
# Prediction -> Success rate (Accuracy)
print("AUC: ",(results.mean()), results.std())
```

```python
'''
le seuil de classement ici est de 0.9
'''
```
